```python
# import the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Create data

X= np.linspace(0*10,50)

Y = 0.5*X +np.random.randn(50)

# Add outliers to Y

Y[-1] += 30
Y[-2] += 50

X= np.vstack([np.ones(50),X]).T
X
```

```python
# Plot data
plt.scatter(X,Y)
```

```python
# Create a Linear Regression model

# The line is yanked towards the ouliers

w = np.linalg.solve(X.T.dot(X),X.T.dot(Y))
Yhat = X.dot(w)

plt.scatter(X[:,1],Y)
plt.plot(X[:,1],Yhat)
```

```python
# Use L2 regularisation(Ridge Regression)

I = np.eye(2)
Lambda = 1000
w1= np.linalg.solve(Lambda*I + X.T.dot(X),X.T.dot(Y))

Yhat1 = X.dot(w1)
plt.scatter(X[:,1],Y)
plt.plot(X[:,1],Yhat,c='red',label="No L2 Regularization")
plt.plot(X[:,1],Yhat1,c='green',label="L2 Regularization")
plt.legend()
```

```python
# Use L2 regularisation with different Lambda(hyperparameter) settings

I = np.eye(2)
Lambda = 10000
w1= np.linalg.solve(Lambda*I + X.T.dot(X),X.T.dot(Y))

Yhat1 = X.dot(w1)
plt.scatter(X[:,1],Y)
plt.plot(X[:,1],Yhat,c='red',label="No L2 Regularization")
plt.plot(X[:,1],Yhat1,c='green',label="L2 Regularization")
plt.legend()

```
